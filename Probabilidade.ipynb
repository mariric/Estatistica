# Probabilidade

A possibilidade ou o quão provável um evento aleatório pode acontecer.

**Frequência:** contagem de ocorrência dentro de um intervalo.

**Tipos de variáveis**

Algébricas: expressão que contém números e letras, separado por operações matemáticas.

Aleatória: o valor é sujeito a variações, assumindo diversas fórmulas. Tem dois tipos, sendo a discreta(separados e finitos, inteiros) e contínua(infinitos, pode ser decimal).



**Distribuição Normal ou Gaussiana:**

Quanto maior a curva, mais disperso os dados estão da média. Média, mediana e moda possuem o mesmo valor. Eventos aleatórios seguem um padrão.

Quanto mais alargado a curva, mais dispersos estão os dados. Mais estreito, os ados estão mais próximos.

**Teorema do Limite Central**

A média das amostras será muito próxima da média da população, tendendo a ter uma distribuição normal

```python
import pandas as pd
import numpy as np
import random
```

```python
custo = pd.read_csv('/content/Cópia de cost-of-living.csv')
custo.head()
```

```python
# transformar uma coluna numa lista

lista = custo.renda_mensal.values.tolist()
print(lista)
```

# Estimar parâmetros:

Parâmetros populacionais, sendo variáveis de uma população.

**Estimativa Pontual**
Uma única estimativa;

*Estimativa Intervalar**
Estimativa dentro de um intervalo ("entre x e y")

**Intervalo de confiança(margem de erro)**

Tolerância de erro, sendo expressa pela média.

Premissas: precisa ser uma amostragem simples aleatória, população normal, conhecimento do desvio populacional.

CI = X +- Z.s/√n

** Nível de Confiança**
Confiança na média real está dentro da margem de erro.

**Desvio Padrão**

Deverá ser estimada a quantidade de respostas, o quanto este ponto está distante da média ou das demais médias.
Padrão baixo: próximo
Padrão alto: mais disperso

Tamanho da amostra

z² Xp(1-p)/ e² \
1 + (z² Xp(1-p)/ e²N)





```python
import pandas as pd
import numpy as np

from scipy.stats import norm
import scipy.stats as st

import random
```

```python
import numpy as np
import scipy.stats as st

# Dados
samples = [12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 25, 27, 28, 33, 34]

# Intervalo de confiança de 95%
teste_95 = st.t.interval(confidence=0.95, df=len(samples)-1, loc=np.mean(samples), scale=st.sem(samples))

print("Intervalo de confiança (95%):", teste_95)
```

```python
# amostra sistemática (salto de 05 em 05 elementos)

def amostragem_sistematica(df,step):

  indexes = np.arange(5, len(df), step=step)
  amostragem_sistematica = df.iloc[indexes]
  return amostragem_sistematica
amostragem_sistematica(movies, 5)
```
