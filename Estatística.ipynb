# Amostragem

** amostra enviesada:** quando um tipo de indivíduo da população tem maior ou menor chance de aparecer na amostra, tendo dificuldade de distribuição da amostra.

** Erro amostral:** quando não é selecionado uma amostra que representa a população alvo.

#Amostragem sistêmica Teórica

É uma variação da amostragem simples, onde após a identificação do elemento, é determinado um critério.

```python
import pandas as pd
import numpy as np
```

```python
iris_df = pd.read_csv('/content/Iris.csv')
iris_df.head(10)
```

```python
iris_df.shape
```

```python
# amostra sistêmica
# criar uma função para criar uma amostra sistematica

def amostragem_sistematica(df, step): # verificar de quantos em quantos ele vai "pular"
  indexes = np.arange(0, len(df), step=step)
  amostra_sistematica = df.iloc[indexes]
  return amostra_sistematica

amostragem_sistematica = amostragem_sistematica(iris_df, 3)
amostragem_sistematica.head(20)
```

# Amostragem estratificada

A população é dividida em subgrupos em função de uma característica em comum, sendo chamado de estado. Após feito isso, cada elemento recebe uma identificação no estrato.

- Viés reduzido;
- Não pode ser usada em todos os estudos, por ser obrigatória a classificação dos subgrupos.

```python
import pandas as pd
import numpy as np
```

```python
iris_df = pd.read_csv('/content/Iris.csv')
iris_df.head(10)
```

```python
# amostragem estratificada

# selecionar dois elementos aletórios de cada subgrupo

iris_df.groupby('Species',group_keys=False).apply(lambda x: x.sample(4))

# verificar uma distribuição igualitária, tarzendo a mesma quantidade de elementos a todos os grupos
```

# Amostragem clusterizada

A unidade de amostragem são os próprios clusters, enquanto na estratificada a amostragem são os elementos de uma população.

- Estudar característica de uma população muito grande;
- Redução de custos com a estratificação;
- Risco maior de erro amostral;
Requer tamanho preciso dos subgrupos

```python
import pandas as pd
import numpy as np
```

```python
iris_df = pd.read_csv('/content/Iris.csv')
iris_df.head(10)
```

```python
# !. Criar cluster através de parâmetros

def amostra_cluster(df, n_por_cluster, num_cluster):
  N = len(df)
  K = int(N/n_por_cluster)
  data = None
  for k in range(K):
    sample_k = df.sample(n_por_cluster)
    sample_k['cluster'] = np.repeat(k, len(sample_k))
    df = df.drop(index=sample_k.index)
    data = pd.concat([data, sample_k], axis=0)


  escolha_cluster = np.random.randint(0, K, size= num_cluster)
  sample_cluster = data[data.cluster.isin(escolha_cluster)]
  return sample_cluster

sample_cluster = amostra_cluster(df = iris_df, n_por_cluster=20, num_cluster=3)
sample_cluster.head(20)
```

```python
sample_cluster.groupby(['cluster'])['cluster'].count()
```
